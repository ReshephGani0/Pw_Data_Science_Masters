#### Q5. How are DataFrame.size() and DataFrame.shape() different?

**Pandas DataFrame size() :**
The size property is used to get an int representing the number of elements in this object and Return the number of rows if Series. Otherwise, return the number of rows times the number of columns if DataFrame.


Pandas df.size Syntax
> Syntax: dataframe.size
> * Return : Returns size of dataframe/series which is equivalent to total number of elements. That is rows x columns.

**Pandas DataFrame shape() :**
The shape property is used to get a tuple representing the dimensionality of the Pandas DataFrame.

Pandas df.shape Syntax
> Syntax: dataframe.shape
> * Return : Returns tuple of shape (Rows, columns) of dataframe/series

```python
data = {'A':['rohan','mohan','sohan','gohan'],
       'B':[24,34,45,23],
        'C':[903830,3483873,34364938,48736548756]}
df=pd.DataFrame(data)
print(df)
print('size : ',df.size)
print('shape : ',df.shape)
```

#### Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
data= {'Name' : ['rohan','mohan', 'sohan', 'gohan','goku','sfd','dfwww'],
       'Values' : [10,20,30,31,96,40,50]}

data2= {'Name' : ['rohan','mohan', 'sohan', 'gohan','goku','sfd'],
       'Values' : [1,2,3,8,4,5]}


df = pd.DataFrame(data2)
df

# creating methods to calculate mean, median and standard deviation
def find_means(x):
    return sum(x)/len(x)
    
def find_medians(x):
    #x.sort()
    x.sort_values()
    lent = len(x)
    if lent%2==0:
        median1= x[lent//2]
        median2 = x[lent//2 -1]
        return (median1+median2)/2
    else:
        return x[lent//2]
    
    
def find_sample_stand_deviate(lst): #sample std is divide by :/n-1 ,but population std is divided by : /n only
    # where n is number of elements in list
    m = sum(lst)/len(lst)
    return ( sum( [(i-m)**2  for i in lst] ) / (len(lst)-1) )**0.5
    
    
df['Values']
print("Dataframe's Values column mean is : ",find_means(df['Values']))

print("Dataframe's Values column median is : ",find_medians(df['Values']))

print("Dataframe's Values column mode is : ", find_sample_stand_deviate(df['Values']))
#(sum([(i-find_means(ls))**2  for i in ls])/(len(ls)-1))**0.5
    
```
